#import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

#define the time axis(from -5 to 5 with a fine resolution)
#we use a high resoultion(0.01)for smooth continues plots
t=np.arange(-5,5,0.01)


#generate the signals
unit_step=(t>=0).astype(float) #step:0 before t=0,1 after
unit_impulse=(t==0).astype(float) #impulse:1 only at t=0
ramp=t*(t>=0).astype(float) #ramp:0before t=0,then t
sine_wave=np.sin(2*np.pi*1*t) #sine:1 hz frequency
exponential=np.exp(-0.5*t)*(t>=0).astype(float) #exponential:e^

#plot everything using subplots
fig,axs=plt.subplots(5,1,figsize=(10,10))

#unit step
axs[0].plot(t,unit_step,linewidth=2,color="b")
axs[0].set_title("unit step signal")
axs[0].grid(True)


# Unit Impulse Signal
# For impulses, we use stem plots instead of line plots

axs[1].stem(t, unit_impulse,
            linefmt='r-',
            markerfmt='ro',
            basefmt='')

axs[1].set_title('Unit Impulse Signal')
axs[1].grid(True)

#ramp
axs[2].plot(t,ramp,linewidth=2,color="g")
axs[2].set_title("ramp signal")
axs[2].grid(True)

#sine wave
axs[3].plot(t,sine_wave,linewidth=2,color="m")
axs[3].set_title("sinusoidal signal(1hz)")
axs[3].grid(True)

#exponential decay
axs[4].plot(t,exponential,linewidth=2,color="r")
axs[4].set_title("exponential decay signal")
axs[4].grid(True)
plt.tight_layout()
plt.show()
