```python
#for loop-iteraring over a range
print('for loop counting 1 to 5:')
for i in range(1,6):
  print(f"count:{i}")
```

```python

#for loop-iterating over a list
print("\nfor loop-iterating through fruits:")
fruits=["apple","banana","orange","grape"]
for fruit in fruits:
  print(f"i like{fruit}")
```

```python

#for loop-iterating over string
print("\nfor loop-iterating through character:")
word="python"
for letter in word:
  print(f"letter:{letter}")
```

```python

# for -while loop
print("\nwhile loop-countdown:")

count = 5

while count > 0:
    print(f"countdown:{count}")
    count -= 1

print("blast off!")
```

```python

# break and continue example
print("\nbreak and continue example")

for num in range(1, 11):
    if num == 5:
        continue  # Skip 5

    if num == 8:
        break     # Stop loop at 8

    print(f"Number: {num}")
```

```python

#creating lists
numbers =[1,2,3,4,5]
mixed=[1,"hello",3.14,True]
colors=["red","green","blue"]

print(f"Numbers: {numbers}")
print(f"Mixed list: {mixed}")
print(f"Colors: {colors}")
```

```python

# Accessing elements
print(f"\nFirst color: {colors[0]}")
print(f"Last color: {colors[-1]}")
```

```python

# List operations
colors.append("yellow")     # Add to end
print(f"After append: {colors}")
```

```python

colors.insert(1, "orange")  # Insert at position
print(f"After insert: {colors}")
```

```python
colors.remove("green")       # Remove element
print(f"After remove: {colors}")
```

```python

#slicing list
print(f"\n first two colors: {colors[:2]}")
print(f"last two colors: {colors[-2:]}")
```

```python

#list comprehension
squares=[x**2 for x in range(1,6)]
print(f"\nsquares: {squares}")
```
